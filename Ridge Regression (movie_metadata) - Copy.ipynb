```python
# drop rows with missing gross revenue data
movie_data=data.dropna(subset=['gross'])
movie_data
```

```python
# only keep  data on movies from USA
movie_data=movie_data[movie_data['country']=='USA']
movie_data
```

```python
movie_data.shape
```

```python
#Assuming a average USA inflation rate of 2.5% we convert gross revenue in terms of 2017 dillars. (100% + 2.5% = 1.025)
movie_data[['gross']]=(1.025**(2017-movie_data['title_year']))*movie_data['gross']
movie_data
```

```python
#Only keep the variable of intrest ,'imdb_score' and 'gross'
movie_data=movie_data[['gross','imdb_score']]
movie_data
```

```python
#let's scale thr gross revenue to be in millions of dollars so its easier tp read
movie_data[['gross']]=movie_data[['gross']]/1000000
movie_data
```

```python
cross_val_scores_ridge=[]
```

```python
alpha=[]
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
from statistics import mean
for i in range(1,9):
    ridgemodel=Ridge(alpha=i*0.50)
    ridgemodel.fit(x_train,y_train)
    score=cross_val_score(ridgemodel,x,y,cv=10)
    avg_cross_val_score=mean(score)*100
    cross_val_scores_ridge.append(avg_cross_val_score)
    alpha.append(i*0.50)
```

```python
# Loop to print the different values of cross-validation scores 
for i  in range (0 , len(alpha)):
    print(str(alpha[i])+' : '+str(cross_val_scores_ridge[i]))
```

```python
# Building and fitting the Ridge Regression model 
ridgemodelchoosen=Ridge(alpha=2)
ridgemodelchoosen.fit(x_train,y_train)
# Evaluating the Ridge Regression model 
print(ridgemodelchoosen.score(x_train,y_train))
```
